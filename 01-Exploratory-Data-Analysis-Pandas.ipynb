```python
df = pd.read_csv('data/taxis_bikes_nycity.csv', parse_dates = ['Data'])
# Carregar novamentes os dados, para mudar o comportamento da 'Data'
# Incluindo o PARÂMETRO , parse_dates da FUNÇÃO read_csv e indicando Pansa a considerar do tipo data
```

```python
df['Data'].head() # De string o tipo passa a ter datetime
```

```python
#Definindo a coluna 'Data', como sendo índice do DataFrame df
#Util para analisar os dados ao longo do tempo, tendo a data como índice
df.set_index('Data', inplace = True)
```

```python
df.head(10)
# A coluna data como ÍNDICE e 'Distância' e 'Tempo' como sendo coluna de DADOS
# Podemos agora realizar uma série de fatiamentos, montar serie temporal para acompanhar os dados ao longo do tempo
# Muito mais possibilidades de manipulação
# Esta conversão anterior depende do objetivo do projeto
```

```python
df.plot() # Criar um plot a partir do DataFrame
# Automaticamente reconhece que o ÍNDICE de df é data. 
# Já cria plotando automaticamente a série temporal 
```

```python
?df.plot #Se não definir os parâmetros, os padrões rodarão 
```

```python
df.plot(kind = 'bar') # Alterar para gráfico tipo de Barras
plt.show() # Printar Gráfico 
```

```python
df.plot(kind = 'area') # Plotar gráfico de área mudando parâmetro da função plot
plt.ylabel("Count")    # Atenção aos picos para não atrapalhar no processo de modelage preditiva
```
