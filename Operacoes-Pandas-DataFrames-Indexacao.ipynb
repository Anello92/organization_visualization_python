## Manipulando e Organizando DataFrames

```python
colunas = list(range(26))
```

```python
ucase = [x for x in string.ascii_uppercase]
lcase = [x for x in string.ascii_lowercase]
```

```python
alfabeto = pd.DataFrame([lcase, ucase, colunas])
```

```python
alfabeto
```

```python
# Transposta da Matriz (DataFrame)
alfabeto = alfabeto.T
```

```python
alfabeto.head()
```

```python
alfabeto.index
```

```python
alfabeto.columns
```

```python
alfabeto.columns = ['minusculo', 'maiusculo', 'numero']
```

```python
alfabeto
```

```python
alfabeto.minusculo
```

```python
alfabeto['minusculo']
```

```python
alfabeto.index = pd.date_range('7/1/2019', periods = 26)
```

```python
alfabeto
```

```python
alfabeto['2019-07-12':'2019-07-19']
```
